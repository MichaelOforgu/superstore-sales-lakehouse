```python
# Create fact_order_details gold table
spark.sql(f"""
CREATE TABLE IF NOT EXISTS {gold_catalog}.fact_order_details AS
SELECT
    od.order_id,
    od.order_detail_id,
    ca.item_key,
    od.amount,
    od.unit_price,
    od.total_price
FROM {silver_catalog}.order_details_clean od
LEFT JOIN {gold_catalog}.dim_categories ca
    ON od.item_id = ca.item_id
""")
```
