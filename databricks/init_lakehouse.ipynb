```python
"""
Initialize lakehouse structure for Sales domain
"""

# Create catalog
spark.sql(f"""
CREATE CATALOG IF NOT EXISTS {env}
COMMENT 'Superstore Sales data lakehouse'
""")

# Use catalog
spark.sql(f"""
USE CATALOG {env}
""")

# Create schemas
spark.sql(f"""
CREATE SCHEMA IF NOT EXISTS {bronze_catalog}
""")

spark.sql(f"""
CREATE SCHEMA IF NOT EXISTS {silver_catalog}
""")

spark.sql(f"""
CREATE SCHEMA IF NOT EXISTS {gold_catalog}
""")

# Create external volume for raw ingestion
spark.sql(f"""
CREATE EXTERNAL VOLUME IF NOT EXISTS {bronze_catalog}.bronze_volume
LOCATION 'abfss://bronze@stsuperstoredevnech.dfs.core.windows.net/'
""")
```
