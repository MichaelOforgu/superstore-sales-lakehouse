```python
# Create Product Performance View
spark.sql(f"""
CREATE OR REPLACE VIEW {gold_catalog}.v_product_performance AS
SELECT
    category1,
    category2,
    item_name,
    SUM(amount) AS quantity_sold,
    ROUND(SUM(total_price),2) AS revenue
FROM {gold_catalog}.v_sales
GROUP BY
    category1,
    category2,
    item_name
""")
```
