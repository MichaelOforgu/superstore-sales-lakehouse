```python
# Create Main Sales View
spark.sql(f"""
CREATE OR REPLACE VIEW {gold_catalog}.v_sales AS
SELECT
    o.order_id,
    o.order_date,
    cu.full_name AS customer_name,
    br.town AS branch_name,
    br.branch_town,
    ca.item_name,
    ca.category1,
    ca.category2,
    ca.brand,
    od.amount,
    od.unit_price,
    od.total_price
FROM {gold_catalog}.fact_orders o
INNER JOIN {gold_catalog}.fact_order_details od
    ON o.order_id = od.order_id
LEFT JOIN {gold_catalog}.dim_customers cu
    ON o.customer_key = cu.customer_key
LEFT JOIN {gold_catalog}.dim_branches br
    ON o.branch_key = br.branch_key
LEFT JOIN {gold_catalog}.dim_categories ca
    ON od.item_key = ca.item_key
""")
```
