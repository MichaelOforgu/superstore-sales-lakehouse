```python
# Create Branch Performance View
spark.sql(f"""
CREATE OR REPLACE VIEW {gold_catalog}.v_branch_sales AS
SELECT
    branch_name,
    branch_town,
    COUNT(DISTINCT order_id) AS orders,
    ROUND(SUM(total_price),2) AS revenue
FROM {gold_catalog}.v_sales
GROUP BY branch_name, branch_town       
""")
```
