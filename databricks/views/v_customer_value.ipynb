```python
# Create Customer Value View
spark.sql(f"""
CREATE OR REPLACE VIEW {gold_catalog}.v_customer_value AS
SELECT
    customer_name,
    COUNT(DISTINCT order_id) AS orders,
    ROUND(SUM(total_price),2) AS lifetime_value
FROM {gold_catalog}.v_sales
GROUP BY customer_name
""")
```
